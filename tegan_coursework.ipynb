```python
# for when we print the images later
def plot_chars(instances, images_per_row=10, **options):
    size = 20
    images_per_row = min(len(instances), images_per_row)
    images = [ndimage.rotate((np.array(instance).reshape(size,size)),270,reshape=True) for instance in instances]
    n_rows = (len(instances) - 1) // images_per_row + 1
    row_images = []
    n_empty = n_rows * images_per_row - len(instances)
    images.append(np.zeros((size, size * n_empty)))
    for row in range(n_rows):
        rimages = images[row * images_per_row : (row + 1) * images_per_row]
        row_images.append(np.concatenate(rimages, axis=1))
    image = np.concatenate(row_images, axis=0)
    plt.imshow(image, cmap = mpl.cm.binary, **options)
    plt.axis("off")
```

### Finding Most Correlating Features per Class

Using SelectBestK to choose the 200, 500 and 1000 best features per class, and then adding them to the corresponding sub dataset.

dataset1 = top 100 features per class = 100*5 = 500 features total

dataset2 = top 200 features per class = 200*5 = 1000 features total

dataset3 = top 300 features per class = 300*5 = 1500 features total

```python
# making new dataframes to store the values of the subsets of the dataset
dataset1_values = []
dataset2_values = []
dataset3_values = []
feature_names = list(range(400))

for character in list(range(5)):
    df_for_character = data[data['Class']==character]
    #print(df_for_class_i)
    X_i = df_for_character.iloc[:,:-1]
    y_i = df_for_character.iloc[:, -1]
    
    # top 100 features
    test = SelectKBest(score_func=chi2, k=100)
    fit = test.fit(X_i, y_i)
    mask = fit.get_support() #list of booleans
    for bool, feature in zip(mask, feature_names):
        if bool:
            dataset1_values.append(feature)
    
    # top 200 features
    test = SelectKBest(score_func=chi2, k=200)
    fit = test.fit(X_i, y_i)
    mask = fit.get_support() #list of booleans
    for bool, feature in zip(mask, feature_names):
        if bool:
            dataset2_values.append(feature)
    
    # top 300 features
    test = SelectKBest(score_func=chi2, k=300)
    fit = test.fit(X_i, y_i)
    mask = fit.get_support() #list of booleans
    for bool, feature in zip(mask, feature_names):
        if bool:
            dataset3_values.append(feature)
    
    
classes = pd.DataFrame(y)
dataset1_values = list(set(dataset1_values))
dataset2_values = list(set(dataset2_values))
dataset3_values = list(set(dataset3_values))
dataset1 = data[dataset1_values]
dataset2 = data[dataset2_values]
dataset3 = data[dataset3_values]
print("Size of dataset1: "+str(len(dataset1.columns)))
print("Size of dataset2: "+str(len(dataset2.columns)))
print("Size of dataset3: "+str(len(dataset3.columns)))
```

```python
def linear_assignment(cost_matrix):
  try:
    import lap
    _, x, y = lap.lapjv(cost_matrix, extend_cost=True)
    return np.array([[y[i], i] for i in x if i >= 0])
  except ImportError:
    from scipy.optimize import linear_sum_assignment
    x, y = linear_sum_assignment(cost_matrix)
    return np.array(list(zip(x, y)))
```

### Varying the Number of Clusters with KMeans

```python
cluster_numbers = [1, 2,3, 4,5, 7, 9, 11, 16]
cluster_accuracys = []
cluster_inertias = []
cluster_homogeneitys = []
```

```python
for k in cluster_numbers:
    kmeans = KMeans(n_clusters = k,random_state=88)
    X_transformed = pca.fit_transform(X_transformed)
    labels = kmeans.fit_predict(X_transformed)
    cm = confusion_matrix(y, labels)
    score = np.trace(cm) / np.sum(cm)
    cluster_accuracys.append(score)
    inertia = kmeans.inertia_
    cluster_inertias.append(inertia)
    homogeneity = metrics.homogeneity_score(y,kmeans.labels_)
    cluster_homogeneitys.append(homogeneity)
    print("Number of clusters: "+str(k))
    print("Accuracy: "+str(score))
    print("Inertia: "+str(inertia))
    print("Homogeneity: "+str(homogeneity))
    print()
```

- Homogenity: A perfectly homogeneous clustering is one where each cluster has data-points belonging to the same class label. Homogeneity describes the closeness of the clustering algorithm to this perfection.

```python
plt.plot(cluster_numbers, cluster_homogeneitys, '-ok')
plt.xlabel("K")
plt.ylabel("Homogeneity")
plt.show()
```

```python
plt.plot(cluster_numbers, cluster_accuracys, '-ok')
plt.xlabel("K")
plt.ylabel("Accuracy Score")
plt.show()
```

### Elbow Method for finding optimum number of clusters for KMeans

- Inertia measures how well a dataset was clustered by K-Means. It is calculated by measuring the distance between each data point and its centroid, squaring this distance, and summing these squares across one cluster. A good model is one with low inertia AND a low number of clusters ( K ).

```python
plt.plot(cluster_numbers, cluster_inertias, '-ok')
plt.xlabel("K")
plt.ylabel("Inertia")
plt.show()
```

### Mean Shift (Hierarchical Clustering) 

- Mean shift clustering using a flat kernel. Mean shift clustering aims to discover “blobs” in a smooth density of samples. It is a centroid-based algorithm, which works by updating candidates for centroids to be the mean of the points within a given region. ... If not set, the seeds are calculated by clustering.

```python
from sklearn.cluster import MeanShift
bandwidths = [1,3,5,10]
meanshift_accuracys = []

for b in bandwidths:
    meanshift = MeanShift(bandwidth=b)
    labels = meanshift.fit_predict(X)
    cm = confusion_matrix(y, labels)
    accuracy = np.trace(cm) / np.sum(cm)
    meanshift_accuracys.append(accuracy)

plt.plot(bandwidths, meanshift_accuracys, '-ok')
plt.xlabel("Bandwidth")
plt.ylabel("Accuracy Score")
plt.show()
```

```python
meanshift_acc = meanshift_accuracys[2]
print("Mean Shift Accuracy: "+str(meanshift_acc))
```

Findings:
- Mean Shift Clustering is much slower than KMeans clustering.
- Accuracy is lower.

### Mini Batch KMeans

- The Mini-batch K-means clustering algorithm is a version of the standard K-means algorithm in machine learning. It uses small, random, fixed-size batches of data to store in memory, and then with each iteration, a random sample of the data is collected and used to update the clusters.

```python
from sklearn.cluster import MiniBatchKMeans
clusters = [3, 5, 7, 9, 11, 13, 15]
kmeans_accuracys = []

for k in clusters:
    kmeans = MiniBatchKMeans(n_clusters=k,random_state=88)
    labels = kmeans.fit_predict(X)
    cm = confusion_matrix(y, kmeans.labels_)
    accuracy = np.trace(cm) / np.sum(cm)
    kmeans_accuracys.append(accuracy)

plt.plot(clusters, kmeans_accuracys, '-ok')
plt.xlabel("No of clusters")
plt.ylabel("Accuracy Score")
plt.show()
```

```python
minikmeans_acc = kmeans_accuracys[1]
print("Mini KMeans Accuracy: "+str(minikmeans_acc))
```

Findings:

- Accuracy is lower than with KMeans.
